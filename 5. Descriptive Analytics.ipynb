```python
# Now we will create a function to count the occurrences of 'None', 'Private', and 'Public' in the 'admin' column
def count_admin_types(row):
    """
    Count the occurrences of 'None', 'Private', and 'Public' in the 'admin' column of a DataFrame.

    Args:
    row (pd.Series): A row of a DataFrame containing an 'admin' column with comma-separated values.

    Returns:
    pd.Series: A Series containing counts of 'None', 'Private', and 'Public' occurrences.
    """
    # Split the string by comma and strip whitespace
    types = row['admin'].split(',')
    types = [t.strip() for t in types]
    
    # Count the occurrences
    none_count = types.count('None')
    private_count = types.count('Private')
    public_count = types.count('Public')
    
    return pd.Series([none_count, private_count, public_count], index=['None', 'Private', 'Public'])

# Apply the function to each row in the dataframe
counts_df = merged_df.apply(count_admin_types, axis=1)

# Concatenate the new counts to the original dataframe
merged_df = pd.concat([merged_df, counts_df], axis=1)

# Create the corrected pivot table
pivot_medical_centers_admin = merged_df.pivot_table(index='neighborhood_name', values=['None', 'Private', 'Public'], aggfunc='sum')

# Display the pivot table
pivot_medical_centers_admin
```

## Index Creation

I will create an Index for medical centers per capita (per population) and per net density (hab/ha). By neighborhood.

```python
# Splitting the 'medical_center' column into lists and counting the number of medical centers
merged_df['medical_center_count'] = merged_df['medical_center'].apply(lambda x: len(x.split(',')) if not pd.isna(x) else 0)

# Creating the index for medical centers per population
merged_df['med_centers_per_pop'] = merged_df['medical_center_count'] / merged_df['population']

# Creating the index for medical centers per net density
merged_df['med_centers_per_density'] = merged_df['medical_center_count'] / merged_df['net_density_hab_ha']
```

#### Medical centers per capita:

For instance, a neighborhood with a smaller population but a large number of healthcare facilities might actually have better access to healthcare on a per capita basis than a larger neighborhood with the same number of facilities.

```python
# Medical centers per density by neighborhood
merged_df[['med_centers_per_density']]
```

#### Medical centers per density:

The measure of "total medical centers divided by net density (hab/ha)" provides a perspective on the availability of medical centers in relation to the population density of an area. Essentially, this metric adjusts the number of medical centers by the population density in each neighborhood.

For example, a higher value in this measure could indicate that there is a relatively high number of medical centers available for a smaller number of people per hectare, suggesting better access to healthcare per person in less densely populated areas. Conversely, a lower value could indicate that there are fewer medical centers available for a larger number of people per hectare, reflecting lower healthcare availability in more densely populated areas.

### Add this columns to the merged_df

```python
# First, we'll define a function to clean and count the medical centers
def count_medical_centers(medical_centers_str):
    """
    Count the number of medical centers from a comma-separated string.

    Args:
    medical_centers_str (str): A string containing comma-separated medical center names.

    Returns:
    int: The count of medical centers in the string.
    """
    # Split the string by comma and filter out 'None' or empty strings
    medical_centers_list = [mc.strip() for mc in medical_centers_str.split(',') if mc.strip() and mc.strip() != 'None']
    
    # Return the count
    return len(medical_centers_list)

# Apply this function to the 'medical_center' column to create a new 'medical_center_count' column
merged_df['medical_center_count'] = merged_df['medical_center'].apply(count_medical_centers)

# Now calculate 'med_centers_per_pop'
merged_df['med_centers_per_pop'] = merged_df['medical_center_count'] / merged_df['population']

# And 'med_centers_per_density'
merged_df['med_centers_per_density'] = merged_df['medical_center_count'] / merged_df['net_density_hab_ha']


merged_df.head(20)
```

### Grouping neighborhoods by income quartiles and analyzing their relationship with the availability of medical centers

```python
# 'admin' column with comma-separated values

# Create a new DataFrame to store the modified 'admin' values
modified_df = merged_df.copy()

# Split the 'admin' column into separate rows in the new DataFrame
s = merged_df['admin'].str.split(', ').apply(pd.Series, 1).stack()
s.index = s.index.droplevel(-1) # to line up with merged_df's index
s.name = 'admin' # needs a name to join

# Remove the original 'admin' column from the new DataFrame and join the new one
del modified_df['admin']
modified_df = modified_df.join(s)

# Group by 'income_quartile' and 'admin' to get counts
pivot_table = modified_df.groupby(['income_quartile', 'admin']).size().unstack(fill_value=0)

# Plot the stacked bar chart
pivot_table.plot(kind='bar', stacked=True)
```

#### Insights:

**1.** As income increases across quartiles, the total number of medical centers (public and private combined) also increases. This is evident in the significant jump in total medical centers from Q3 to Q4.

**2.** There is a large proportion of private medical centers to public ones across all income quartiles, indicating that private medical centers are more prevalent regardless of the income level of the neighborhood.

**3.** The highest income quartile (Q4) not only has the most medical centers in total but also the highest number of public medical centers compared to other quartiles.

This information could suggest that while higher income areas may have better access to medical facilities overall, the presence of private medical centers is a common factor across all levels of neighborhood income. It also seems that public medical facilities are not evenly distributed, with higher-income areas having more. This aligns with our hypothesis that wealthier neighborhoods have more medical resources. However, for a more thorough analysis, it would be important to consider the size and capacity of these centers as well, not just the count.

### Distribution of Medical Centers by Neighborhood and Income (Sorted by Ascending Median Income)
