### Detect Outliers Using IQR Method and Boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=numerical_data, color='lightblue')
plt.title('Boxplot of Gross Income')
plt.xlabel('Gross Income (EUR)')
plt.show()
```

#### Outliers:

Data points outside the "whiskers" of the box plot are considered outliers. These are values that are above or below 1.5 times the IQR from the upper and lower quartiles, respectively.
There is a significant number of outliers above the third quartile, indicating that there are individuals with much higher incomes than the majority.

#### Extreme Values:

Extreme values beyond 100,000 EUR could represent individuals with exceptionally high incomes.

#### Justification for Not Removing Outliers

#### Real Income Distribution:

In a real income context, it is entirely plausible that there are individuals or households with much higher incomes than the average. Removing these outliers could distort the true understanding of the income distribution.

#### Relevance for Analysis:

Outliers can be critical for certain analyses, such as income inequality studies, where high incomes are a fundamental part of the research topic.

#### Representation of Subpopulations:

Outliers may represent subpopulations within Barcelona, such as high-level executives... or districts known for affluence.

#### No Evidence of Errors:

If there are no errors in data collection or entry, and these high values reflect real situations, there is no statistical justification for removing them from the analysis.

#### Additional Considerations

#### Outlier-Sensitive Analysis:

In this context the outliers are relevant to the research question, so they should be retained in the dataset. We could use methods to reduce the influence of outliers without necessarily removing them.

#### Data Transformation:

If outliers negatively affect the analysis, we may consider data transformation, such as applying a logarithm or square root, to reduce the impact of extreme values rather than eliminating them.

In summary, **outliers in the context of income can be valid data points that reveal the true economic structure of a city. The decision to keep them should be based on an understanding of the data context and the objectives of the analysis. So I will not remove them.**

```python
def detect_outliers_iqr(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

outliers = detect_outliers_iqr(numerical_data)
print("Outliers:", numerical_data[outliers])
```

```python
# Detect outliers in numerical_data and print the count of those outliers
outliers = detect_outliers_iqr(numerical_data)
outlier_count = np.sum(outliers)
print("Number of Outliers:", outlier_count)
```

```python
# Calculate the percentage of outliers in a dataset and print the result
outliers = detect_outliers_iqr(numerical_data)
outlier_count = np.sum(outliers)
total_data_points = numerical_data.shape[0]
percentage_outliers = (outlier_count / total_data_points) * 100
print("Percentage of Outliers:", percentage_outliers, "%")
```

Approximately 7.12% of the data points in the 'gross_income_eur' column are classified as outliers based on the Interquartile Range (IQR) method.

### Compare Mean and Median

```python
mean_income = np.mean(numerical_data)
median_income = np.median(numerical_data)
difference = mean_income - median_income
print("Mean Income:", mean_income)
print("Median Income:", median_income)
print("Difference (Mean - Median):", difference)
```

#### Important Insights:

- Mean Income: The mean income for the population is approximately 53,840.53 EUR.

- Median Income: The median income is 49,079.50 EUR.

The difference between the mean and median income is approximately 4,761.03 EUR.

As I said previously, the presence of outliers and the difference between the mean and median income suggest that the income distribution is skewed to the right, meaning there are relatively few individuals or households with very high incomes that are pulling the mean income upwards.

#### Justification for Using Median Income for Analysis by Neighborhoods:

Given the presence of outliers and the skewed distribution, it is appropriate to **use the median income as a measure of central tendency when analyzing income by neighborhoods**. **The median is less influenced by extreme values (outliers) than the mean and provides a more robust representation of the typical income in each neighborhood**. This approach ensures that the analysis is not unduly affected by a small number of extremely high-income individuals or households, which may not be representative of the majority of the population in each neighborhood.

## Creating a neighborhood_median_income

There is no column that indicates the median income for each neighborhood. So, I will create a new column named neighborhood_median_income that will represent the median income for each neighboorhod.

```python
# Group by 'neighborhood_name' and calculate the median income for each neighborhood
neighborhood_median_income = income_df.groupby('neighborhood_name')['gross_income_eur'].median().reset_index()

# Rename the columns for clarity
neighborhood_median_income.columns = ['neighborhood_name', 'median_income']

# Merge the median income data back into the original income_df
income_df = pd.merge(income_df, neighborhood_median_income, on='neighborhood_name')

# Now, income_df contains the 'median_income' column associated with each neighborhood
```
