## Create new columns: type and admin

### type column

I will create a new column called 'type' in the medical_centers_df DataFrame, which classifies the type of medical center based on the 'medical_center' column. The classify_center_type function checks if the string contains keywords that identify whether it's a Hospital, a CAP (Centre d'Atenció Primària in catalan = Primary Care Clinic), or another type of center. If the name is None or an empty string, the function returns 'None'.

```python
# We will create a function to classify the type of medical center and apply that function to create the new 'type' column.

# Define the function to classify the center type based on keywords
def classify_center_type(name):
    # Check if the value is None or an empty string
    if name in [None, 'None', '']:
        return 'None'
    name_lower = name.lower()
    if 'hospital' in name_lower:
        return 'Hospital'
    elif "centre d'atenció primària" in name_lower:
        return 'CAP'
    elif "centre d'urgències d'atenció primària" in name_lower:
        return 'CUAP'
    else:
        return 'Other'

# Apply the function to create the new 'type' column
medical_centers_df['type'] = medical_centers_df['medical_center'].apply(classify_center_type)

medical_centers_df.head()  # Show the first rows to verify the results
```

```python
medical_centers_df
```

### admin column

To classify the medical centers as public or private, I relied on the information available on the website "https://guia.barcelona.cat/en/". In Barcelona, all CAP (Primary Care Centers) and CUAP (Urgent Primary Care Centers) are public. Among the hospitals listed, the following are public:

- Hospital Clínic i Provincial
- Hospital del Mar
- Hospital de l'Esperança
- Vall d'Hebron Barcelona Hospital Campus

All other medical centers are considered private.

```python
def classify_administration(center_name, center_type):
    """
    Classifies whether a medical center is public or private based on the provided rules.

    Args:
    center_name (str): The name of the medical center.
    center_type (str): The type of the medical center.

    Returns:
    str: 'Public' if the medical center is classified as public, 'Private' if private, or 'None' for other cases.

    Rules:
    - If 'center_type' is 'CAP' or 'CUAP', it is considered public.
    - If 'center_type' is 'Hospital' and 'center_name' is in the list of public hospitals, it is considered public.
    - For all other cases (including None and Other), it is considered private.

    """
    
    # List of public hospitals based on their names
    public_hospitals = [
        'Hospital Clínic i Provincial', 
        'Hospital del Mar', 'Hospital de l\'Esperança', "Vall d'Hebron Barcelona Hospital Campus"
    ]
    
    # If 'center_type' is CAP or CUAP, it's public
    if center_type in ['CAP', 'CUAP']:
        return 'Public'
    
    # If 'center_type' is Hospital and the center name is in the list of public hospitals, it's public
    if center_type == 'Hospital' and center_name in public_hospitals:
        return 'Public'   
    
    # For all other cases (including None and Other), it's private
    else:
        return 'Private'


# Apply the function to create the 'admin' column
medical_centers_df['admin'] = medical_centers_df.apply(
    lambda row: classify_administration(row['medical_center'], row['type']) if row['medical_center'] is not None and row['medical_center'] != 'None' else 'None', 
    axis=1
)

medical_centers_df.head()  # Show the first rows to verify the results
```

## Group and Sort by neighborhood_name

```python
# Sort the DataFrame by 'neighborhood_name'
medical_centers_df = medical_centers_df.sort_values(by='neighborhood_name')

# Define a custom aggregation function to concatenate values into a single string separated by commas while preserving order
def custom_agg(series):
    return ', '.join(series.astype(str))

# Group by 'neighborhood_name' and aggregate columns using the custom_agg function
aggregations = {
    'medical_center': custom_agg,
    'year': 'first',
    'district_name': 'first',
    'type': custom_agg,
    'admin': custom_agg
}

# Reset the index to reflect the new order
medical_centers_df = medical_centers_df.groupby('neighborhood_name').agg(aggregations).reset_index()
```
