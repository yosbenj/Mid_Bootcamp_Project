# Import Libraries

```python
import pymysql
import getpass # To get the password without showing the input
import sqlalchemy
from sqlalchemy import create_engine

import pandas as pd
import numpy as np
```

### Load the DataFrame from the CSV file

```python
healthcare_df = pd.read_csv('barcelona_healthcare_df.csv')
healthcare_df.head()
```

### Establish a connection between Python and the Project database

```python
password = getpass.getpass("Please enter the SQL database password: ")
```

```python
# Specify the connection to the SQL database
connection_string = 'mysql+pymysql://root:'+password+'@localhost/project'

# Create an SQLAlchemy engine
engine = sqlalchemy.create_engine(connection_string)
```

### Load the DataFrame into the SQL database

```python
healthcare_df.to_sql(name='barcelona_healthcare', con=engine, schema=None, if_exists='replace', index=False)
```

### Read the DataFrame from SQL

```python
# Define SQL query
query = "SELECT * FROM barcelona_healthcare"

# Execute the query and load the results into a DataFrame
barcelona_healthcare_df = pd.read_sql(query, con=engine)
```
